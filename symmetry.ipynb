# Symétrie du LiMgBi (mp-570213)

```python
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter
from IPython.display import FileLink
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
from pymatgen.core.operations import SymmOp
```

```python
key = "1sKG6aj6JUQi8D22lH"  #my key for API access
material = "mp-570213"   #ID of my material
m = MPRester(key)
struct = m.get_structure_by_material_id(material,conventional_unit_cell=True) #Getting structure object
w = CifWriter(struct)    #Write to CIF
w.write_file(material+".cif") 
print("Click on the link to download my CIF file")
display(FileLink(material+".cif"))
```

# Réflexion

```python
Op = SymmOp.from_xyz_string('-y,-x,z') #Création de la symétrie
pos = struct.sites[2].frac_coords #application sur l'atome de Li 2
result = Op.operate(pos)
print ("Position de l'atome de Li : "+ str(pos))
print ("Position de la symétrie : "+ str(result))
```


## Description
La réflexion est une opération qui inverse la chiralité et qui se fait par rapport à un plan miroir qu'on voit en vert. L'équation de la réflexion dans ce cas-ci est : 
$$\begin{bmatrix} 0 & -1 & 0 \\ -1 & 0 & 0 \\ 0 & 0 & 1\end{bmatrix} \begin{bmatrix} x \\ y \\ z\end{bmatrix} = \begin{bmatrix} -y \\ -x \\ z \end{bmatrix} $$


# Rotation

```python
Op = SymmOp.from_xyz_string('x-y,-y,-z') #Création de la symétrie
pos = struct.sites[5].frac_coords #application sur l'atome de Mg 2
result = Op.operate(pos)
print ("Position de l'atome de Mg : "+ str(pos))
print ("Position de la symétrie : "+ str(result))
```


## Description
La rotation est une opération qui tourne l'atome de 180 degrés autour d'un axe (ici le vecteur a). L'équation est dans ce cas-ci :
$$\begin{bmatrix} 1 & -1 & 0 \\ 0 & -1 & 0 \\ 0 & 0 & -1\end{bmatrix} \begin{bmatrix} x \\ y \\ z\end{bmatrix} = \begin{bmatrix} x-y \\ -y \\ -z \end{bmatrix} $$


```python
Op = SymmOp.from_xyz_string('-y,-z,-x') #Création de la symétrie
pos = struct.sites[10].frac_coords #application sur l'atome de Bi10
result = Op.operate(pos)
print ("Position de l'atome de Bi : "+ str(pos))
print ("Position de la symétrie : "+ str(result))
```


## Description
La rotoinversion est une opération qui combine une rotation autour d'un axe, ici l'axe (1 1 1) avec une inversion. L'équation est dans ce cas-ci :
$$\begin{bmatrix} 0 & -1 & 0 \\ 0 & 0 & -1 \\ -1 & 0 & 0\end{bmatrix} \begin{bmatrix} x \\ y \\ z\end{bmatrix} = \begin{bmatrix} -y \\ -z \\ -x \end{bmatrix} $$
